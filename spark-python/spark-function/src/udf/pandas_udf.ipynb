```python
from pyspark.sql.functions import pandas_udf
from pyspark.sql import SparkSession
import pandas as pd
from pyspark.sql.types import StringType
```

```python
spark = SparkSession.builder.appName("UDF Example").getOrCreate()
```

```python
# Sample DataFrame
df = spark.createDataFrame([(1, "john"), (2, "jane")], ["id", "name"])
```

```python
# Define a Pandas UDF
@pandas_udf(StringType())
def capitalize_pandas_udf(names: pd.Series) -> pd.Series:
    return names.str.capitalize()
```

```python
# Apply to DataFrame
df.withColumn("capitalized_name", capitalize_pandas_udf(df["name"])).show()
```

```python
from pyspark.sql.types import IntegerType
from pyspark.sql.functions import struct
```

```python

@pandas_udf(IntegerType())
def add_numbers_pandas(a: pd.Series, b: pd.Series) -> pd.Series:
    return a + b
```

```python
df = spark.createDataFrame([(1, 10), (2, 20)], ["id", "value"])
```

```python
df.withColumn("new_value", add_numbers_pandas(df["id"], df["value"])).show()
```
