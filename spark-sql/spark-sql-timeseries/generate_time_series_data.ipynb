```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Generate Time Series Data

```python
# Parameters
np.random.seed(42)
date_range = pd.date_range(start="2023-01-01", end="2024-01-01", freq="D")
n = len(date_range)

# Generate data
trend = np.linspace(0, 10, n)  # Linear trend
seasonal = 5 * np.sin(np.linspace(0, 2 * np.pi, n))  # Seasonal component
noise = np.random.normal(0, 1, n)  # Random noise

# Combine them to form the time series
data = trend + seasonal + noise

# Create a DataFrame
df = pd.DataFrame({"timestamp": date_range, "value": data})
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df["timestamp"], df["value"], label="Synthetic Time Series Data")
plt.xlabel("Date")
plt.ylabel("Value")
plt.title("Generated Time Series Data")
plt.legend()
plt.show()
```

## Save the Data to a CSV File (Optional)
